```python
# import turkish calendar
t_cal = pd.read_csv("/content/drive/Shareddrives/Projeto II/CSV_raw/Turkish_calendar.csv", sep=';')
t_cal.head()
```

```python
t_cal.info()
```

```python
# Convert the column to datetime with the correct format
t_cal = pd.DataFrame(t_cal)
t_cal.rename(columns={'CALENDAR_DATE': 'date',
                      'DAY_OF_MONTH': 'day',
                      'DAY_OF_WEEK_SK': 'weekday',
                      'SEASON_SK': 'season',
                      'WEEK_OF_YEAR': 'week',
                      'PUBLIC_HOLIDAY_FLAG': 'holiday'}, inplace=True)
t_cal['date'] = pd.to_datetime(t_cal['date'], format='%d.%m.%Y')

t_cal.info()
```

```python
t_cal.head()
```

```python
# Join tables

features = pd.merge(sales, product, on='product_id', how='left')
features = pd.merge(features, cities, on = 'store_id', how='left')
features = pd.merge(features, t_cal[['date', 'day', 'weekday', 'season', 'week', 'holiday']], on = 'date', how='left')

# Reorder the columns
#ordered_columns = ['date', 'city_code','store_id', 'product_id', 'sales', 'revenue', 'stock', 'price']
#remaining_columns = [col for col in features.columns if col not in ordered_columns]
#features = features[ordered_columns + remaining_columns]

features
```

```python
features.info()
```

```python
# drop last month where no sales have been registered

# Define start and end dates
start_date = '2017-01-02'
end_date = '2019-09-30'

features_filtered = features[(features['date'] >= start_date) & (features['date'] <= end_date)]

features_filtered.head()
features_filtered.tail()
```

```python
#Weekdays' Dictionary
week_days = {1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat', 7: 'Sun'}

# Creating the new column with week day names
features_filtered['weekday'] = features_filtered['weekday'].map(week_days)
features_filtered.head()
```

```python
#Month

# Convert the date column to datetime format
features_filtered['date'] = pd.to_datetime(features_filtered['date'])

# Extract the abbreviated month name and create a new column
features_filtered['month_name'] = features_filtered['date'].dt.strftime('%b')
```

```python
features_filtered.head()
```

```python
# confirm drop of last month

max_date = features_filtered['date'].max()
max_date
```
